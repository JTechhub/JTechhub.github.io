#**4. 분석과 시각화를 위한 tidy data 만들기**

```python
df_raw.columns
```

```python
df = df_raw.melt(id_vars=["지역별", "상품군별", "판매유형별", "항목", "단위"],
            var_name="기간", value_name="백만원")
df.shape
```

```python
df.head()
```

```python
df.info()
```

# **5. 데이터 전처리**


###**5-1. 기간에서 연도 분리하기**

```python
df["기간"]
```

```python
df["연도"] = df["기간"].str.extract(r"(\d{4})\.\d+/\d+")

df.head()
```

```python
#기간에서 분기만 분리하기
df["분기"] = df["기간"].str.extract(r"(\d+)/\d+")

df.head()
```

###**5-2. 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기**

```python
# - 문자를 결측치로 변경하고 float 타입으로 변경하기
df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float)
```

###**5-3. 필요없는 데이터 제거하기**

```python
#합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거합니다.
#2016
df= df[(df["지역별"] != "합계") & (df["상품군별"] != "합계")].copy()
df.info()
```

```python
#결측치 보기
df.isnull().sum()
```

#**6. 데이터 시각화**

###**6-1. 전체 상품군 판매액**

```python
#판매유형별 데이터는 일부 기간에는 "계"만 존재하기 때문에
#판매유형별 == "계" 데이터만 가져서 봐야지
#평균 값을 구하는 그래프에서 올바른 값을 표현할 수 있습니다.
df_total = df[df["판매유형별"] == "계"].copy()
df_total.head()
```

```python
#연도, 판매액 lineplot 그리기
sns.lineplot(data = df_total, x = "연도", y = "백만원")
```

```python
#연도, 판매액 lineplot으로 그리고 상품군별로 다른 색상으로 표시하기
sns.lineplot(data = df_total, x = "연도", y = "백만원", hue = "상품군별")
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```

```python
#위에 그린 그래프를 자세히 보기 위해 서브플롯으로 표시하기
sns.relplot(data = df_total, x = "연도", y = "백만원",
            hue = "상품군별", kind = "line", col = "상품군별", col_wrap = 4)
```

```python
#isin을 사용해 화장품만 제외를 하고 df_sub 라는 변수에 담습니다.
df_sub = df_total[-df_total["상품군별"].isin(["화장품", "의류 및 패션 관련 상품"])].copy()
```

###**6-2. 화장품의 온라인쇼핑 해외직접판매액**

```python
#df_cosmetic 이라는 변수에 상품군별이 화장품인 데이터만 가져오기
df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic["상품군별"].unique()
```

```python
#연도와 판매액을 lineplot으로 그리고 분기별로 다른 색상으로 표현해보기
plt.figure(figsize=(15,4))
sns.lineplot(data = df_cosmetic, x = "연도", y = "백만원", hue = "분기")
```

```python
#화장품 판매액에 대한 기간별 금액 데이터 시각화 하기
plt.figure(figsize=(15,4))
plt.xticks(rotation = 30)
sns.lineplot(data = df_cosmetic, x = "기간", y = "백만원")
```

```python
#화장품 판매액에 대한 기간별 금액 데이터 시각화하고 판매유형별로 다른 색상 표시하기
plt.figure(figsize=(15, 4))
plt.xticks(rotation=30)
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy()
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별")
```

```python
#화장품 판매액에 대한 기간별 금액 데이터 시각화하고 지역별로 다른 색상 표시하기
plt.figure(figsize=(15,10))
plt.xticks(rotation = 30)
sns.lineplot(data = df_cosmetic[df_cosmetic["지역별"] !="중국"], x = "기간", y = "백만원", hue = "지역별")
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```

###**6-3. 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액**

```python
#df_fashion 이라는 변수에 의류 데이터만 가져와 따로 담아주기
df_fashion = df[(df["상품군별"] == "의류 및 패션 관련 상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
#의류 및 패션 관련 상품 판매액에 대한 기간별 금액 데이터 시각화하고
#지역별로 다른색상으로 표시하기
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_fashion, x = "기간", y = "백만원", hue = "지역별")
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```

```python
#의류 및 패션 관련 상품 판매액에 대한 기간별 금액 데이터 시각화하고
#판매유형별로 다른색상으로 표시하기
df_fashion2 = df[(df["상품군별"] == "의류 및 패션 관련 상품") & (df["판매유형별"] != "계") & (df["판매유형별"] != "면세점")].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_fashion2, x = "기간", y = "백만원", hue = "판매유형별", errorbar=None)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```

###**6-4. 데이터 집계하기**

```python
#피봇테이블로 지역별, 연도별 합계 금액을 표 형태로 구하기
result = df_cosmetic.pivot_table(index = "지역별", columns = "연도",
                        values = "백만원", aggfunc = "sum")
result
```

```python
#피봇테이블로 지역별, 연도별 합계 금액을 표 형태로 구하기
result2 = df_fashion.pivot_table(index = "지역별", columns = "연도",
                        values = "백만원", aggfunc = "sum")
result2
```

###**6-5. 연산결과를 시각적으로 보기**

```python
#피벗테이블로 구한 결과를 값의 많고 적음에 따라 시각적으로 표현하기
plt.figure(figsize=(10,6))
sns.heatmap(result[result.index != "중국"], cmap = "Blues", annot = True, fmt = ".0f")
```

```python
#피벗테이블로 구한 결과를 값의 많고 적음에 따라 시각적으로 표현하기
plt.figure(figsize=(10,6))
sns.heatmap(result2, cmap = "Blues", annot = True, fmt = ".0f")
```
